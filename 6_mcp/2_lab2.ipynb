### Now we write an MCP server and use it directly!

```python
# Now let's use our accounts server as an MCP server

params = {"command": "uv", "args": ["run", "accounts_server.py"]}
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as server:
    mcp_tools = await server.list_tools()
```

```python
mcp_tools
```

```python
instructions = "You are able to manage an account for a client, and answer questions about the account."
request = "My name is Ed and my account is under the name Ed. What's my balance and my holdings?"
model = "gpt-4.1-mini"
```

```python
from openai import AsyncOpenAI
from agents import Agent, OpenAIChatCompletionsModel, Runner, set_tracing_disabled
import os
import mlflow

# Enable auto tracing for OpenAI Agents SDK
mlflow.openai.autolog()

# Optional: Set a tracking URI and an experiment
mlflow.set_tracking_uri("http://localhost:5555")
mlflow.set_experiment("OpenAI Agent")


BASE_URL = "http://localhost:11434/v1"
client = AsyncOpenAI(base_url=BASE_URL)
MODEL_NAME = "qwen3:latest"
set_tracing_disabled(disabled=True)


async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as mcp_server:
    agent = Agent(
        name="account_manager", 
        instructions=instructions, 
        model=OpenAIChatCompletionsModel(model=MODEL_NAME, openai_client=client), 
        mcp_servers=[mcp_server]
        )
    with trace("account_manager"):
        result = await Runner.run(agent, request)
    display(Markdown(result.final_output))
```

```python
from IPython.display import display, Markdown
from agents import Agent, Runner, OpenAIChatCompletionsModel, set_tracing_disabled
from agents.mcp import MCPServerStdio
from openai import AsyncOpenAI
import mlflow

# Enable auto tracing for OpenAI Agents SDK
mlflow.openai.autolog()

# Optional: Set a tracking URI and an experiment
mlflow.set_tracking_uri("http://localhost:5555")
mlflow.set_experiment("OpenAI Agent")

BASE_URL = "http://localhost:11434/v1"
client = AsyncOpenAI(base_url=BASE_URL)

# Disable tracing
set_tracing_disabled(disabled=True)

params = {"command": "uv", "args": ["run", "calculator_server.py"]}

# Use async directly in notebook
async with MCPServerStdio(params=params, client_session_timeout_seconds=30) as my_server:
    print("Server started!")
    agent = Agent(
        name="My calculator", 
        instructions="You are a helpful math assistant",
        model=OpenAIChatCompletionsModel(model="qwen3:latest", openai_client=client),
        mcp_servers=[my_server]
    )
    result = await Runner.run(agent, "Adding these numbers 1, 40, 96 and divide it by 4")
    display(Markdown(result.final_output))
```
